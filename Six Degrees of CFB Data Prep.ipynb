```python
import numpy as np
import pandas as pd
import networkx as nx
```

```python
#read in data gathered from https://borderingonwisdom.wordpress.com/college-football-coach-database/
df = pd.read_excel(r'C:\Users\BD\Football-Coach-Database.xlsx', sheet_name = 'History by Head Coach')
df
```

```python
#remove unwanted rows
df = df[df['Current Head Coaches'] != 'Former Head Coaches']
df
```

```python
#remove unneeded columns
df = df.drop(['Age','Birthdate','Alma Mater'], axis = 1)
df
```

```python
#create variables needed for melt (de-pivoting of data)
id_vars = ['Current Head Coaches']
value_vars = df[df.drop('Current Head Coaches', axis = 1).columns] #list of all columns but 'Current Head Coaches'
var_name = 'year'
value_name = 'team'
```

```python
#collapse data from pivot table into a database-like structure
df = df.melt(id_vars=id_vars, value_vars=value_vars, var_name=var_name, value_name=value_name)
df
```

```python
#remove blanks (years where a coach did not coach)
df = df.dropna()
df
```

```python
#rename columns for ease of use
df.columns = ['coach','year','team']
df
```

```python
#test data to eye check validity
df[df['coach'] == 'Mike Norvell']
```

```python
#use str.split to separate the team and position values (e.g. separate 'Arizona State - OC' into 'Arizona State', 'OC')
df[['team', 'position']] = df['team'].str.split(" - ",  expand = True)
df
```

```python
#another eye check
df[df['coach'] == 'Mike Norvell']
```

```python
#when no position is listed, that means the coach was Head Coach. Fill in the blanks to reflect that
df['position'] = df['position'].fillna('HC')
df
```

```python
#create another column combining the year and team into a single variable
#this allows us to easily distinguish between different iterations of the same team when we create the network
df['squad'] = df['year'].apply(str) + ' '+ df['team']
df
```

```python
df.to_csv('CFB Coaches.csv', index = False)
```
